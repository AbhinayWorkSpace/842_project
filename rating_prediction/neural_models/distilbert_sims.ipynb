```python
import pickle

file_path = '/content/drive/MyDrive/distilbert_binary14.pkl'

with open(file_path, 'rb') as f:
    distil_bin14 = pickle.load(f)

distil_bin14
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, accuracy_score, f1_score
import matplotlib.pyplot as plt
import numpy as np

cm = confusion_matrix(distil_bin14['actual_labels'], distil_bin14['predicted_labels'])
fig, ax = plt.subplots(figsize=(5, 5))
label_map = {0: "1.0", 1:"4.0"}
cmd = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[label_map[i] for i in range(len(label_map))])
cmd.plot(cmap='Blues', ax=ax)

plt.title("DistilBERT Confusion Matrix")
plt.show()

accuracy = accuracy_score(distil_bin14['actual_labels'], distil_bin14['predicted_labels'])
precision = precision_score(distil_bin14['actual_labels'], distil_bin14['predicted_labels'], average='micro')
recall = recall_score(distil_bin14['actual_labels'], distil_bin14['predicted_labels'], average='micro')
f1 = f1_score(distil_bin14['actual_labels'], distil_bin14['predicted_labels'], average='micro')

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```
