```python
import json
import pandas as pd

with open(file_path1, 'r') as file:
    data = [json.loads(line) for line in file]

with open(file_path2, 'r') as file:
    for line in file:
      data.append(json.loads(line))

df = pd.DataFrame(data)
df
```

```python
df = df.loc[df['verified_purchase'] != False]
df = df.reset_index(drop=True)
df
```

```python
rating_counts = df['rating'].value_counts()
min_reviews = rating_counts.min()

df = pd.concat([df[df['rating'] == rating].sample(min_reviews) for rating in range(1, 6)])
df = df.sample(frac=1).reset_index(drop=True)

df['rating'].value_counts()
```

```python
from transformers import TFBertForSequenceClassification, BertTokenizer
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

ratings = df['rating'].tolist()

label_encoder = LabelEncoder()
ratings_encoded = label_encoder.fit_transform(ratings)

X_train, X_test, y_train, y_test = train_test_split(df["text"].tolist(), ratings_encoded, test_size=0.2)
```

```python
from transformers import AutoTokenizer
import numpy as np

tokenizer = AutoTokenizer.from_pretrained('roberta-base')

train_data = tokenizer(X_train, return_tensors="np", padding=True, truncation=True, max_length=512)
train_data = dict(train_data)

test_data = tokenizer(X_test, return_tensors="np", padding=True, truncation=True, max_length=512)
test_data = dict(test_data)
```

```python
train_data
```

```python
from keras import backend

backend.clear_session()
```

```python
from transformers import TFAutoModelForSequenceClassification
import numpy as np

model = TFAutoModelForSequenceClassification.from_pretrained('roberta-base', num_labels=len(label_encoder.classes_))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_data,
    y=np.array(y_train),
    epochs=1,
    batch_size=4
)
```

```python
evaluation = model.evaluate(
    test_data,
    y=np.array(y_test),
    batch_size=4
)
```
