```python
import pickle

file_path = '/content/drive/MyDrive/roberta_binary.pkl'

with open(file_path, 'rb') as f:
    rob_bin = pickle.load(f)

rob_bin
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, accuracy_score, f1_score
import matplotlib.pyplot as plt
import numpy as np

cm = confusion_matrix(rob_bin['actual_labels'], rob_bin['predicted_labels'])
fig, ax = plt.subplots(figsize=(5, 5))
label_map = {0: "1.0", 1: "5.0"}
cmd = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[label_map[i] for i in range(len(label_map))])
cmd.plot(cmap='Blues', ax=ax)

plt.title("RoBERTa Confusion Matrix")
plt.show()

accuracy = accuracy_score(rob_bin['actual_labels'], rob_bin['predicted_labels'])
precision = precision_score(rob_bin['actual_labels'], rob_bin['predicted_labels'], average='micro')
recall = recall_score(rob_bin['actual_labels'], rob_bin['predicted_labels'], average='micro')
f1 = f1_score(rob_bin['actual_labels'], rob_bin['predicted_labels'], average='micro')

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```
