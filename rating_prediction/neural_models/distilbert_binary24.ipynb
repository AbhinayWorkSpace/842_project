```python
import json
import pandas as pd

with open(file_path1, 'r') as file:
    data = [json.loads(line) for line in file]

with open(file_path2, 'r') as file:
    for line in file:
      data.append(json.loads(line))

df = pd.DataFrame(data)
df
```

```python
df = df.loc[df['verified_purchase'] != False]
df = df.reset_index(drop=True)
df
```

```python
rating_counts = df['rating'].value_counts()
min_reviews = rating_counts.min()

df = pd.concat([df[df['rating'] == rating].sample(min_reviews) for rating in range(1, 6)])
df = df.sample(frac=1).reset_index(drop=True)

df['rating'].value_counts()
```

```python
df = df[df['rating'].isin([2.0, 4.0])]
df
```

```python
df['rating'].value_counts()
```

```python
(df['title'] + ' ' + df['text']).iloc[0]
```

```python
from transformers import TFBertForSequenceClassification, BertTokenizer
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

ratings = df['rating'].tolist()

label_encoder = LabelEncoder()
ratings_encoded = label_encoder.fit_transform(ratings)

X_train, X_test, y_train, y_test = train_test_split((df['title'] + ' ' + df['text']).tolist(), ratings_encoded, test_size=0.2)
```

```python
from transformers import AutoTokenizer
import numpy as np

tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')

train_data = tokenizer(X_train, return_tensors="tf", padding=True, truncation=True, max_length=512)
train_data = dict(train_data)

test_data = tokenizer(X_test, return_tensors="tf", padding=True, truncation=True, max_length=512)
test_data = dict(test_data)
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    train_data,
    y_train
)).batch(16)

test_dataset = tf.data.Dataset.from_tensor_slices((
    test_data,
    y_test
)).batch(16)
```

```python
from keras import backend

backend.clear_session()
```

```python
from transformers import TFAutoModelForSequenceClassification
import numpy as np

model = TFAutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=len(label_encoder.classes_))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_dataset,
    epochs=1
)
```

```python
evaluation = model.evaluate(test_dataset)
```

```python
history.history
```

```python
evaluation
```

```python
predictions = model.predict(test_dataset)
```

```python
predicted_labels = np.argmax(predictions.logits, axis=1)
predicted_labels
```

```python
import pickle

file_path = '/content/drive/MyDrive/distilbert_binary24.pkl'

with open(file_path, 'wb') as f:
    pickle.dump({
        'evaluation': evaluation,
        'history': history.history,
        'predicted_labels': predicted_labels,
        'actual_labels': y_test
        }, f)
```
