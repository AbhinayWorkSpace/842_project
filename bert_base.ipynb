```python
import os

if os.path.exists(file_path):
    print("File exists!")
else:
    print("File does not exist.")
```

```python
import json
import pandas as pd

with open(file_path, 'r') as file:
    data = [json.loads(line) for line in file]

df = pd.DataFrame(data)
df
```

```python
df = df.loc[df['verified_purchase'] != False]
df = df.reset_index(drop=True)
df
```

```python
!pip install tensorflow-text
```

```python
df = df[:10000]
df
```

```python
import tensorflow as tf
from transformers import TFBertTokenizer, TFBertForSequenceClassification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

reviews = df['text'].tolist()
ratings = df['rating'].tolist()
```

```python
label_encoder = LabelEncoder()
ratings_encoded = label_encoder.fit_transform(ratings)

ratings_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(reviews, ratings_encoded, test_size=0.2)
```

```python
tokenizer = TFBertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(label_encoder.classes_))
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding='max_length', max_length=128)
test_encodings = tokenizer(X_test, truncation=True, padding='max_length', max_length=128)
```

```python
train_encodings
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
)).shuffle(1000).batch(16)

train_dataset
```

```python
test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
)).batch(16)

test_dataset
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(train_dataset, epochs=3)
```

```python
model.evaluate(test_dataset)
```
